```python
t = Symbol('t')
x = Symbol('x')

#tutaj wpisujecie swoje zadanie (x' = ...)
def f(t,x):
    return t - 2*t*x
```

```python
def f_t(t,x):
    return diff(f(t,x), t)

def f_x(t,x):
    return diff(f(t,x), x)

def f_tx(t,x):
    return diff(f(t,x), t, 1, x, 1)

def f_tt(t,x):
    return diff(f(t,x), t, 2)

def f_xx(t,x):
    return diff(f(t,x), x, 2)
```

```python
def F1(t,x):
    return f_t(t,x)+f(t,x)*f_x(t,x)
 
F1(t,x)
```

```python
def F2(t,x):
    return f_tt(t,x)+2*f(t,x)*f_tx(t,x)+f_t(t,x)*f_x(t,x)+f(t,x)*(f_x(t,x))**2+(f(t,x))**2*f_xx(t,x)

F2(t,x)
```

```python
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.integrate import odeint
```

```python
#warunek z tresci zadania
x0 = -3               
init = x0 
def rown(x, t):
    return t - 2*t*x
```

```python
#przedział z treści zadania
t1 = np.linspace(0, 1, 11)
t2 = np.linspace(0, 1, 21)
t3 = np.linspace(0, 1, 101)

sol_d1 = odeint(rown, init, t1)

H = [0.1, 0.05, 0.01]
```

```python
# TAYLOR

def taylora(f, f1, f2,  x0, t, h):
    n = len(t)
    x = np.zeros((n, len(x0)))
    x[0] = x0
    for i in range(n - 1):
        k = f(x[i], t[i])
        k1 = f1(x[i], t[i])
        k2 = f2(x[i], t[i])
        x[i+1] = x[i] + k*h + 1/2*k1*h**2 + 1/6*k2*h**3
    return x

#funkcja ma zwracać to co wypisuje 5 komórka
def F1(x,t):
    return -2*t*(-2*t*x+t)-2*x+1

#funkcja ma zwracać co wypisuje 6 komórka
def F2(x,t):
    return (4*t**2)*(-2*t*x+t)+8*t*x-2*t*(1-2*x)-4*t

#wyniki dla różnych wartości h
sol00 = taylora(rown, F1, F2, [x0], t1, H[0])
sol01 = taylora(rown, F1, F2, [x0], t2, H[1])
sol02 = taylora(rown, F1, F2, [x0], t3, H[2])
#print(sol01)
#print(sol01[::2, 0])
#print(sol01[:, ::2])

print(len(t2),len(sol_d1[:, 0]),len(sol01[::2, 0]),len(sol_d1[:, 0]),len(sol01[::2, 0]))
#zmiencie tutaj nazwę kolumny 'blad_globalny,', żeby to tak samo nie wygladało
th1 = pd.DataFrame(data={'t': t1, 'dokładne': sol_d1[:, 0], 'taylora': sol00[:,0], 'blad_globalny': abs(sol_d1[:, 0] - sol00[:, 0])})
th2 = pd.DataFrame(data={'t': t2[::2], 'dokładne': sol_d1[:, 0], 'taylora': sol01[::2, 0], 'blad_globalny': abs(sol_d1[:, 0] - sol01[::2, 0])})
th3 = pd.DataFrame(data={'t': t3[::10], 'dokładne': sol_d1[:, 0], 'taylora': sol02[::10, 0], 'blad_globalny': abs(sol_d1[:, 0] - sol02[::10, 0])})


```

```python
#EULER
def Eulera(f, x0, t, h):
    n = len(t)
    x = np.zeros((n, len(x0)))
    x[0] = x0
    for i in range(n - 1):
        x[i+1] = x[i] + h* (t[i+1] - t[i]) * f(x[i], t[i])
    return x
 
sol10 = Eulera(rown, [x0], t1, H[0])
sol11 = Eulera(rown, [x0], t2, H[1])
sol12 = Eulera(rown, [x0], t3, H[2])

eh1 = pd.DataFrame(data={'t': t1, 'dokładne': sol_d1[:, 0], 'eulera': sol10[:, 0], 'blad_globalny': abs(sol_d1[:, 0] - sol10[:, 0])})
eh2 = pd.DataFrame(data={'t': t2[::2], 'dokładne': sol_d1[:, 0], 'eulera': sol11[::2, 0], 'blad_globalny': abs(sol_d1[:, 0] - sol11[::2, 0])})
eh3 = pd.DataFrame(data={'t': t3[::10], 'dokładne': sol_d1[:, 0], 'eulera': sol12[::10, 0], 'blad_globalny': abs(sol_d1[:, 0] - sol12[::10, 0])})

```

```python
#MIDPOINT
def midpoint(f, x0, t, h):
    n = len(t)
    x = np.zeros((n, len(x0)))
    x[0] = x0
    for i in range(n - 1):
        k1 = f(x[i], t[i])
        k2 = f(x[i] + k1 * h / 2., t[i] + h / 2.)
        x[i+1] = x[i] + k2*h
    return x

sol20 = midpoint(rown, [x0], t1, H[0])
sol21 = midpoint(rown, [x0], t2, H[1])
sol22 = midpoint(rown, [x0], t3, H[2])

mh1 = pd.DataFrame(data={'t': t1, 'dokładne': sol_d1[:, 0], 'midpoint': sol20[:, 0], 'blad_globalny': abs(sol_d1[:, 0] - sol20[:, 0])})
mh2 = pd.DataFrame(data={'t': t2[::2], 'dokładne': sol_d1[:, 0], 'midpoint': sol21[::2, 0], 'blad_globalny': abs(sol_d1[:, 0] - sol21[::2, 0])})
mh3 = pd.DataFrame(data={'t': t3[::10], 'dokładne': sol_d1[:, 0], 'midpoint': sol22[::10, 0], 'blad_globalny': abs(sol_d1[:, 0] - sol22[::10, 0])}) 

mh1

```

```python
#RUNGE KUTTA
def rungekutta4(f, x0, t, h):
    n = len(t)
    x = np.zeros((n, len(x0)))
    x[0] = x0
    for i in range(n - 1):
        k1 = f(x[i], t[i])
        k2 = f(x[i] + k1 * h / 2., t[i] + h / 2.)
        k3 = f(x[i] + k2 * h / 2., t[i] + h / 2.)
        k4 = f(x[i] + k3 * h, t[i] + h)
        x[i+1] = x[i] + (h / 6.) * (k1 + 2*k2 + 2*k3 + k4)
    return x

sol30 = rungekutta4(rown, [x0], t1, H[0])
sol31 = rungekutta4(rown, [x0], t2, H[1])
sol32 = rungekutta4(rown, [x0], t3, H[2])

rh1 = pd.DataFrame(data={'t': t1, 'dokładne': sol_d1[:, 0], 'rungekutta': sol30[:, 0], 'blad_globalny': abs(sol_d1[:, 0] - sol30[:, 0])})
rh2 = pd.DataFrame(data={'t': t2[::2], 'dokładne': sol_d1[:, 0], 'rungekutta': sol31[::2, 0], 'blad_globalny': abs(sol_d1[:, 0] - sol31[::2, 0])})
rh3 = pd.DataFrame(data={'t': t3[::10], 'dokładne': sol_d1[:, 0], 'rungekutta': sol32[::10, 0], 'blad_globalny': abs(sol_d1[:, 0] - sol32[::10, 0])}) 

rh3
```

```python
taylor = pd.DataFrame()
taylor = pd.concat([th1,th2,th3], axis=1)

euler = pd.DataFrame()
euler = pd.concat([eh1,eh2,eh3], axis=1)

rk = pd.DataFrame()
rk = pd.concat([rh1,rh2,rh3], axis=1)

midpoint = pd.DataFrame()
midpoint = pd.concat([mh1,mh2,mh3], axis=1)

with pd.ExcelWriter("rurki.xlsx") as writer:
    taylor.to_excel(writer, sheet_name='taylor')
    euler.to_excel(writer, sheet_name='euler')
    rk.to_excel(writer, sheet_name='rk')
    midpoint.to_excel(writer, sheet_name='mid')
```
